#### Forest fires over the years

```python
df.groupby(["Year"]).sum().sort_values(by="Value", ascending=False).head(30)
```

#### Forst fire area in coutries over years

```python
df.groupby(["Area"]).sum().sort_values(by="Value", ascending=False).head(30)
```

#### Forst type burned over time

```python
df.groupby(["Item"]).sum().sort_values(by="Value", ascending=False).head(30)
```

#### The maximum forest area burned in a year in a country

```python
df.groupby(["Year","Area"]).sum().sort_values(by="Value", ascending=False).head(30)
```

#### Forest type burned in a year

```python
df.groupby(["Year","Item"]).sum().sort_values(by="Value", ascending=False).head(30)
```

```python
df.groupby(["Year","Area","Item"]).sum().sort_values(by="Value", ascending=False).head(30)
```

```python
import plotly.express as px
fig = px.choropleth(df, locations="Area Code (ISO3)", color="Value", hover_name="Area", animation_frame="Year", range_color=[20,80])
fig.show()
```
